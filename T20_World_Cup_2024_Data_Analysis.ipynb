```python
cric_dataset = cricket_dataset.copy()
cric_dataset.head()
```

```python
categorical_dataset = cric_dataset.drop(['First Innings Score','Fall of wickets First Innings',
                                         'Second Innings Score', 'Fall of wickets Second Innings',
                                         'Winning Margin', 'Highest Score', 'Best Bowler Figure(Wickets Taken)',
                                         'Best Bowler Figure(Runs Recieved)'], axis=1)
```

```python
categorical_dataset.head()
```

```python
wrong_type = cric_dataset[['First Innings Score','Fall of wickets First Innings',
                           'Second Innings Score','Fall of wickets Second Innings',
                           'Winning Margin','Highest Score', 'Best Bowler Figure(Wickets Taken)',
                           'Best Bowler Figure(Runs Recieved)']]
```

```python
numeric_Data_Frame = wrong_type.apply(pd.to_numeric, errors='coerce')

numeric_Data_Frame.dtypes
```

```python
numeric_Data_Frame.head()
```

```python
numeric_Data_Frame.describe()
```

```python
# Concat Categorical and Casting Numerical part

cric_dataset = pd.concat([categorical_dataset, numeric_Data_Frame], axis=1)
cric_dataset.head()
```

```python
cric_dataset.info()
```

```python
cric_dataset.columns
```

```python
venue = cric_dataset.groupby(['Venue']).size().reset_index(name='Matches')
venue
```

```python
ply.histogram(data_frame = venue, x='Venue', y = 'Matches', color_discrete_sequence=['#36256E'],text_auto = True)
```

### Interpretation:
Most Matches:
We can see the venue where most of the matches of the T20 World Cup played was 'Kensington Oval' which is located in Bridgetown, Barbados. A total of 9 matches were played there during T20 World Cup 2024.

There were 2 venues which ranked second in terms of most matches played during T20 World Cup 2024 these venues are 'Nassau County International Cricket Stadium, New York' and 'Sir Vivian Richards Stadium, North Sound, Antigua'. A total of 8 matches were played here.

Least Matches:
We can clearly see that there are two venues where the least number of matches were played during the T20 World Cup 2024 and those venues are 'Central Broward Regional Park Stadium Turf Ground, Lauderhill' and 'Grand Prairie Stadium, Dallas'. Both of these stadium got 4-4 matches to conduct.

### Most wins based on toss win

```python
toss_decision_vs_win = cric_dataset.groupby(['Toss Decision','Winners'])['Toss Decision'].size().reset_index(name='Win')

toss_decision_win_count = toss_decision_vs_win.groupby(['Toss Decision'])['Win'].sum().reset_index(name='Win')
toss_decision_win_count
```

```python
ply.pie(data_frame = toss_decision_win_count, names='Toss Decision', values='Win',
        color_discrete_sequence=['#133A75','#DA2C5A','#1C5D9C'], title='Toss Winning Effect on win')
```

### Interpretation:
We can see that 76.4% of the matches are won by the teams who elected to Field first after winning the toss, whereas 18.2% of the matches are won by the teams who elected to bat first after winning the toss. After all these there were 5.45% of matches that were abandoned and some of them are won by DLS method.

## Distribution of First Innings Score

```python
score = cric_dataset[['First Innings Score','Second Innings Score']]
inning_score = score.dropna().reset_index(drop = True)
inning_score
```

```python
fig = ply.histogram(inning_score,
                    x = 'First Innings Score',
                    marginal='box',
                    nbins=10,
                    title='Distribution of First Innings Score', color_discrete_sequence=['#663399'])
fig.update_layout(bargap=0.4)
fig.show()
```

### Interpretation:

We can see that, the 12 times the first inning score was between 100 and 119, which is the maximum. Whereas there were 9 Matches when the First Inning Score was between 140 and 159.

There were only 3 matches when the score of first inning was between 40 and 59 followed by 3 matche score between 80 and 99.

### Distribution of Second Innings Score

```python
fig1 = ply.histogram(inning_score,
                     x = 'Second Innings Score',
                     marginal='box',
                     nbins = 10,
                     title = 'Distribution of Second Innings Score', color_discrete_sequence=['#800080'])
fig1.update_layout(bargap = 0.4)
fig1.show()
```

### Interpretation:

We can see that 17 times the team which bats 2nd scored the runs between 100 and 119 which is the maximum, whereas 9 times the teams which bats 2nd scored the runs between 120 and 139 which is the second most.
And only one time the team batted 2nd scored the runs between 20 and 39.

### Best Bowling figure in a single match

```python
bowler = cric_dataset[['Best Bowler', 'Best Bowler Figure(Wickets Taken)','Best Bowler Figure(Runs Recieved)']]
```

```python
bowler
```

#### Note:
In this dataset we found an error in 'Best Bowler Figure(Wickets Taken)' columns. It is showing that Nosthush Kenjige got 30 wickets and Alpesh Ramjani got 17 wickets in a single match which are outliers.

```python
best_bowling = bowler.sort_values(by = 'Best Bowler Figure(Wickets Taken)', ascending = False)
best_figure10 = best_bowling.iloc[2:12].reset_index(drop = True)
best_figure10
```

```python
df_melted = best_figure10.melt(id_vars = 'Best Bowler', value_vars=['Best Bowler Figure(Wickets Taken)','Best Bowler Figure(Runs Recieved)'], var_name = 'variable', value_name = 'Value')

fig11 = ply.bar(df_melted, x='Best Bowler', y='Value', facet_col='variable',color_discrete_sequence=['#C71585'], title='Best Bowling Figure')
fig11.show()
```

### Interpretation:
 We can see that the best bowling figure was Fazalhaq Farooqi because he got 5 wickets and conceded 9 runs only.


### Best batsman run figure in a single match

```python
batter = cric_dataset[['Top Scorer','Highest Score']]
top_10_batter = batter.sort_values(by = 'Highest Score', ascending=False)
top_10_batter = top_10_batter[:10].reset_index(drop=True)
top_10_batter
```

```python
sns.barplot(data=top_10_batter, y='Top Scorer',x='Highest Score')
```

### Interpretation:

### Player of the match Through the Tournament

```python
player_of_the_match = cric_dataset.groupby(['Player Of The Match']).size().reset_index(name = 'count')
player_of_the_match = player_of_the_match.nlargest(6, columns='count').reset_index(drop=True)
player_of_the_match
```

```python
top_5_player = player_of_the_match.iloc[1:].reset_index(drop=True)
top_5_player
```

```python
sns.barplot(data = top_5_player, y = 'Player Of The Match', x='count')
plt.title('Top 5 Players with most Man of the Match Awards')
```

### Interpretation:
We can clearly see that Fazalhaq Farooqi has been awarded with player of the match award 3 times which is most by any player. Followed by Adam Zampa, Adil Rashid, Jasprit Bumrah and Marcus Stoinis with 2-2 Man of the Match awards for each.
